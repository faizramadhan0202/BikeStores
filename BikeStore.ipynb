<strong id="Exploratory Data Analysis">Exploratory Data Analysis<strong/>

```python
df['product_name'].value_counts().sort_values(ascending=False)
```

```python
df.groupby('product_name')['total_units'].count().sort_values(ascending=False)
```

```python
df.groupby('product_name')['revenue'].sum().sort_values(ascending=False)
```

```python
#list Category to Plot
cat_vars = ['category_name', 'brand_name', 'store_name', 'sales_rep']
             
#create figure with subplot
fig, axs = plt.subplots(nrows=2, ncols=2, figsize=(30, 30))
axs = axs.ravel()

#create bar plot for each category list
for i, var in enumerate(cat_vars):
    sns.barplot(x=var, y='revenue', data=df, ax=axs[i], estimator=np.mean)
    axs[i].set_xticklabels(axs[i].get_xticklabels(), rotation=90)
    
#adjust spasing between subplots
fig.tight_layout()

#show plot
plt.show()
```

```python
cat_vars = ['revenue', 'total_units']

fig, axs = plt.subplots(nrows=1, ncols=2, figsize=(20,10))

for i, var in enumerate(cat_vars):
    sns.boxplot(x=var, data=df, ax=axs[i])
    
fig.tight_layout()

plt.show()
```

```python
plt.rcParams['figure.figsize'] = (12,6)
sns.countplot(df['state'], palette='pink')
plt.title('City', fontsize=15)
plt.show()
```

```python
plt.figure(figsize = (20, 10))
ax = sns.countplot(x = 'category_name', data=df, palette = 'pastel')
ax.set_title('Count of Category', fontsize=15)

plt.show()
```

```python
plt.figure(figsize = (20, 10))
ax = sns.countplot(x = 'brand_name', data=df, palette = 'pastel')
ax.set_title('Count of Brand', fontsize=15)

plt.show()
```

```python
sns.barplot(y='category_name', x='revenue', data=df)
plt.xticks(rotation=5)
plt.grid(True)
```
